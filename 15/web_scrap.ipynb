### Задание 1

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
URL = 'https://habr.com/ru/search/?target_type='
KEYWORDS = ['python', 'парсинг']
```

```python
def get_all_links(url, query):
    params = {
        'q': query,
    }
    res = requests.get(URL, params)
    soup = BeautifulSoup(res.text, 'html.parser')
    news_blocks = soup.find_all('article', class_='post post_preview')
    articles_intro = list(map(lambda x: x.find('h2', class_='post__title'), news_blocks))
    link_list = list(map(lambda x: x.find('a').get('href'), articles_intro))
    return link_list
```

```python
def get_df_news(links, query):

    df_news = pd.DataFrame()
    for link in links:
        soup = BeautifulSoup(requests.get(link).text, 'html.parser')
        time.sleep(0.3)
        date = soup.find('span', class_='post__time').get_text()
        title = soup.find('span', class_='post__title-text').get_text()
        text = soup.find('div', class_='post__text').get_text()
        row = {'keywords': query, 'date': date, 'title': title, 'link': link, 'text': text}
        df_news = pd.concat([df_news, pd.DataFrame([row])])
    return df_news   
```

```python
def get_result(url, query):
    
    result = pd.DataFrame()
    
    for i in query:        
        news = get_df_news(get_all_links(url, i), i)
        result = pd.concat([news, result])
    
    return result
```

```python
res = get_result(URL, KEYWORDS)
res
```

```python
def data_df(url, email, headers):
    payload = {
            'emailAddresses' : [email]
        }
    r = requests.post(url, json=payload, headers=header)
    df = pd.DataFrame(r.json()['breaches']).T
    df['email'] = email
    return df
```
