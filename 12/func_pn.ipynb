### Задание 1

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
rating_df = ratings[['movieId', 'rating']].groupby('movieId').mean()
```

```python
def classification(row):
    if row.rating<=2:
        return 'низкий рейтинг'
    elif (row.rating>2.0) & (row.rating<=4.0):
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
movies['class'] = rating_df.apply(classification, axis=1)
movies.head(40)
```

```python
requests_df = pd.read_csv('ml-latest-small/keywords.csv')
```

```python
def search(row):
    if ('москва' in row['keyword']) | ('тула' in row['keyword']) | ('ярославль' in row['keyword']):
        return 'Центр'
    elif ('петербург' in row['keyword']) | ('псков' in row['keyword']) | ('мурманск' in row['keyword']):
        return 'Северо-Запад'
    elif('владивосток' in row['keyword']) | ('сахалин' in row['keyword']) | ('хабаровск' in row['keyword']):
        return 'Дальний Восток'
    else:
        return 'undefined'

requests_df['region'] = requests_df.apply(search, axis=1)
requests_df.head()

# проверка
#requests_df.loc[requests_df['region'].isin(['Дальний Восток'])]
```

### Задание 3

```python
import numpy as np
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
#список годов с 1950 по 2020 - задавать по 2010 некорректно, так как есть фильмы с 2011 по 2016 годы выпуска, которые будут записаны в 1900
years = np.arange(1950, 2020)
```

```python
#определяем среднюю оценку каждого фильма
film_rating = ratings[['movieId', 'rating']].groupby('movieId').mean('rating')
```

```python
#функция для поиска года в названии фильма
def production_year(row):
    for i in years:
        if str(i) in row.title:
            return i
    return 1900
```

```python
movies['year'] = movies.apply(production_year, axis=1)
```

```python
#объединяем таблицы названия фильма и рейтинга, вычисляем среднюю оценку по годам, сортируем
joined = movies.merge(film_rating, on='movieId', how='left')
joined[['year', 'rating']].groupby('year').mean().sort_values('rating', ascending=False)
```
